```python
'''
2. Data Preprocessing:
a. Handle missing values (e.g., imputation).
b. Encode categorical variables.
'''
```

```python
#Handle missing values (e.g., imputation).
df.isnull().sum()
```

```python
#Dropping cabin as it has too many missing values
df.drop(columns="Cabin",inplace=True)
```

```python
df
```

```python
#imputing missing age values with median value
df["Age"] = df["Age"].fillna(df["Age"].median())
df
```

```python
#replacing null values in embarked column
df["Embarked"] = df["Embarked"].fillna(method="ffill")
df
```

```python
df.isnull().sum() #null values removed
```

```python
#Encode categorical variables
df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["Sex"] = LE.fit_transform(df["Sex"])
df["Embarked"] = LE.fit_transform(df["Embarked"])
df["Name"] = LE.fit_transform(df["Name"])
df["Ticket"] = LE.fit_transform(df["Ticket"])
df.head()
```

```python
'''
3. Model Building:
a. Build a logistic regression model using appropriate libraries (e.g., scikit-learn).
b. Train the model using the training data.
'''
```

```python
#Build a logistic regression model using appropriate libraries (e.g., scikit-learn).
#split the variables
X_train = df.iloc[:,2:11]
Y_train = df["Survived"]
```

```python
#fit the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred_train = model.predict(X_train)
```

```python
#probability
df["pred_prob"] = model.predict_proba(X_train)[:,1]
```

```python
'''
Model Evaluation:
a. Evaluate the performance of the model on the testing data using accuracy, precision, recall, F1-score, and ROC-AUC score.
Visualize the ROC curve.
'''
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
score = accuracy_score(Y_train,Y_pred_train)
f1score = f1_score(Y_train,Y_pred_train)
precision = precision_score(Y_train,Y_pred_train)
recall = recall_score(Y_train,Y_pred_train)
print("For training data:-")
print("Accuracy score:",round(score,2))
print("F1 score:",round(f1score,2))
print("Precision score:",round(precision,2))
print("Recall score:",round(recall,2))
```

```python
#ROC Curve
from sklearn.metrics import roc_curve,roc_auc_score
fpr,tpr,dummy = roc_curve(Y_train,df["pred_prob"])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=fpr,y=tpr,color="red")
plt.plot(fpr,tpr)
plt.ylabel("True positive Rate")
plt.xlabel("False positive Rate")
plt.show()
```

```python
auc = roc_auc_score(Y_train,df["pred_prob"])
print("Area under curve:", (auc*100).round(2))
```
