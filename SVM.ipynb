```python
'''
Task 2: Data Preprocessing
1.	Encode categorical variables if necessary.
2.	Split the dataset into training and testing sets.
'''
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["class"] = LE.fit_transform(df["class"])
df_label = df["class"]
```

```python
#split the variables
X = df[["stalk_height","cap_diameter"]]
Y = df["class"]
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

```python
df_new = pd.concat([SS_X,df_label],axis=1)
df_new
```

```python
#split the variables
X = df_new[["stalk_height","cap_diameter"]]
Y = df_new["class"]
```

```python
#split the X and Y variables
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 100,test_size=0.2,stratify=Y)
```

```python
'''
Task 3: Data Visualization
1.	Employ scatter plots, pair plots, or relevant visualizations to comprehend feature distributions and relationships.
2.	Visualize class distributions to gauge dataset balance or imbalance.
'''
```

```python
#2D plot
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=Y, cmap='viridis')  # Use Y for color mapping
plt.xlabel("Stalk Height")
plt.ylabel("Cap Diameter")
plt.title("Scatter Plot of Stalk Height vs Cap Diameter")
plt.colorbar(label="Y") # Show colorbar to represent values in Y
plt.show()
```

```python
#3 D visualization

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X['stalk_height'], X['cap_diameter'], Y, c=Y, cmap='viridis')

ax.set_xlabel('Stalk Height')
ax.set_ylabel('Cap Diameter')
ax.set_zlabel('Y')
ax.set_title('3D Scatter Plot of Stalk Height, Cap Diameter, and Class')
plt.show()
```

```python
'''
Task 4: SVM Implementation
1.	Implement a basic SVM classifier using Python libraries like scikit-learn.
2.	Train the SVM model on the training data.
3.	Evaluate model performance on the testing data using appropriate metrics (e.g., accuracy, precision, recall, F1-score).
'''
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel="linear",C=1)
svm_model.fit(X_train,Y_train)
```

```python
Y_pred = svm_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
score = accuracy_score(Y_test,Y_pred)
p_score = precision_score(Y_test,Y_pred)
r_score = recall_score(Y_test,Y_pred)
f1 = f1_score(Y_test,Y_pred)
print("Accuracy score:",round(score,2))
print("Precision score:",round(p_score,2))
print("Recall score:",round(r_score,2))
print("F1 score:",round(f1,2))
```

```python
'''
Task 5: Visualization of SVM Results
1.	Visualize classification results on the testing data.
'''
```

```python
# prompt: visualize the data for the above linear classifer

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
import numpy as np

# Create a meshgrid to visualize the decision boundary
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict the class for each point in the meshgrid
Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and the data points
plt.contourf(xx, yy, Z, alpha=0.4)
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=Y, palette="viridis")
plt.xlabel("Stalk Height")
plt.ylabel("Cap Diameter")
plt.title("Linear SVM Classifier")
plt.show()
```

```python
'''
Task 6: Parameter Tuning and Optimization
1.	Experiment with different SVM hyperparameters (e.g., kernel type, regularization parameter) to optimize performance
'''
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the parameter grid
param_grid = {
    "kernel": ["linear", "poly", "rbf"],
    "C": [0.1, 1, 10],
    "degree": [2, 3],
    "gamma": ["scale", 0.1]
}

# Initialize GridSearchCV
cv_model = GridSearchCV(
    SVC(),
    param_grid=param_grid,
    refit=True,
    verbose=3,
    n_jobs=-1  # Use all available cores
)

# Fit the model
cv_model.fit(X_train, Y_train)

# Predict using the best model
Y_pred = cv_model.predict(X_test)

# Accuracy score
score = accuracy_score(Y_test, Y_pred)
print("Accuracy score:", round(score, 2))

# Show best parameters
print("Best Parameters:", cv_model.best_params_)
```
