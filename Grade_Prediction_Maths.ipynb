```python
num_col = df.select_dtypes(include="number").columns
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df[num_col] = SS.fit_transform(df[num_col])
df.head()
```

```python
cat_col = df.select_dtypes(include="object").columns
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for col in cat_col:
    df[col] = LE.fit_transform(df[col])
df.head()
```

```python
import seaborn as sns
correlation_matrix = df[num_col].corr()
sns.heatmap(correlation_matrix)
```

```python
#split the data
X = df.drop("G3",axis=1)
Y = df["G3"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=46)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_model = GradientBoostingRegressor(n_estimators=100,learning_rate=0.1,random_state=42)
gb_model.fit(X_train,Y_train)
```

```python
Y_pred = gb_model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score
rmse = root_mean_squared_error(Y_test,Y_pred)
r2 = r2_score(Y_test,Y_pred)
print("Root mean squared error:",round(rmse,2))
print("R2 score:",round(r2,2))
```

```python
#cross validation
import numpy as np
rmse = []
r2 = []
for i in range(1,101):
    from sklearn.model_selection import train_test_split
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=i)
    gb_model.fit(X_train,Y_train)
    Y_pred = gb_model.predict(X_test)
    rmse.append(root_mean_squared_error(Y_test,Y_pred))
    r2.append(r2_score(Y_test,Y_pred))
print("Root mean squared error:",round(np.mean(rmse),2))
print("R2 score:",round(np.mean(r2),2))
```

```python
#grid search CV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [2, 3, 4],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Model
gb_model = GradientBoostingRegressor(random_state=42)

# GridSearchCV
gv_model = GridSearchCV(
    estimator=gb_model,
    param_grid=param_grid,
    cv=5,               
    scoring='r2',
    n_jobs=-1            
)

# Fit
gv_model.fit(X_train, Y_train)

# Predictions
Y_pred = gv_model.predict(X_test)

# Evaluation
rmse = root_mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("Best Params:", gv_model.best_params_)
print("Root Mean Squared Error:", round(rmse, 2))
print("R2 Score:", round(r2, 2))
```
