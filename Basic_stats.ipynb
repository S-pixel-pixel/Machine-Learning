```python
'''
Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset
Introduction
●	To perform descriptive analytics, visualize data distributions, and preprocess the dataset for further analysis.
Descriptive Analytics for Numerical Columns
●	Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.
●	Steps:
	Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	Identify numerical columns in the dataset.
	Calculate the mean, median, mode, and standard deviation for these columns.
	Provide a brief interpretation of these statistics.
'''
import pandas as pd
import matplotlib.pyplot as plt
#Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
df = pd.read_csv("C:\\Users\\sujey\\Downloads\\Assignments\\Basic stats - 1\\Basic stats - 1\\sales_data_with_discounts.csv")
df
```

```python
#identify numerical columns
df.select_dtypes(include="number")
```

```python
#Descriptive Statistics
#Calculate the mean, median, mode, and standard deviation for these columns.
round(df.mean(numeric_only=True),2)
```

```python
round(df.median(numeric_only=True),2)
```

```python
round(df.std(numeric_only=True),2)
```

```python
round(df.mode(numeric_only=True),2)
```

```python
'''Data Visualization
●	Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.
●	Histograms:
	Plot histograms for each numerical column.
	Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.'''
# histogram for Avg Price
plt.hist(df['Avg Price'],color='green',edgecolor='black')
plt.title('SALES DATA FOR AVG PRICE',color='black')
plt.xlabel('Average Price',color='black')
plt.ylabel('Frequency',color='black')
plt.show()

# histogram for Volume
plt.hist(df['Volume'],color='blue',edgecolor='black')
plt.title('Quantity of item purchased',color='black')
plt.xlabel('Quantity of item purchased',color='black')
plt.ylabel('Frequency',color='black')
plt.show()

# histogram for Total Sales Value
plt.hist(df['Total Sales Value'],color='grey',edgecolor='black')
plt.title('Sum of total value earned without discount',color='black')
plt.xlabel('total amount earned',color='black')
plt.ylabel('Frequency',color='black')
plt.show()

# histogram for Discount rate(%)
plt.hist(df['Discount Rate (%)'],color='brown',edgecolor='black')
plt.title('Discount Rate (%)',color='black')
plt.xlabel('Discount Rate (%)',color='black')
plt.ylabel('Frequency',color='black')
plt.show()

# histogram for Net Sales Value
plt.hist(df['Net Sales Value'],color='yellow',edgecolor='black')
plt.title('Net Sales Value',color='black')
plt.xlabel('Net Sales Value',color='black')
plt.ylabel('Frequency',color='black')
plt.show()

# histogram for Discount Amount
plt.hist(df['Discount Amount'],color='purple',edgecolor='black')
plt.title('Discount Amount',color='black')
plt.xlabel('Discount Amount',color='black')
plt.ylabel('Frequency',color='black')
plt.show()
```
