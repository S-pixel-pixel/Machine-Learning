```python
#split the variables
X = df.iloc[:,1:]
```

```python
'''
Task 2: Dimensionality Reduction with PCA:
1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of 1 to implement PCA to reduce the dimensionality of the dataset.
2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
3.	Transform the original dataset into the principal components.
'''
#Standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X.head()
```

```python
#PCA
from sklearn.decomposition import PCA
pca = PCA()
PC = pca.fit_transform(SS_X)
PC = pd.DataFrame(PC)
PC.columns = list(X)
PC.head()
```

```python
PC.var()
```

```python
X_new = PC.iloc[:,:5]
```

```python
'''
Task 3: Clustering with Original Data:
1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
2.	Visualize the clustering results using appropriate plots.
3.	Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.
'''
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
score = []
for i in range(2,11):
    cluster = KMeans(n_clusters = i)
    df["K_Means"] = cluster.fit_predict(SS_X)
    score.append(silhouette_score(SS_X,df["K_Means"]))
score
```

```python
#Visualization
import matplotlib.pyplot as plt
plt.scatter(range(2,11),score)
plt.plot(range(2,11),score,color="red")
plt.show()
```

```python
print("The highest silhouette score by using KMeans clustering and without using PCA components is:",round(max(score),2))
```

```python
'''
Task 4: Clustering with PCA Data:
1.	Apply the same clustering algorithm to the PCA-transformed dataset.
2.	Visualize the clustering results obtained from PCA-transformed data.
3.	Compare the clustering results from PCA-transformed data with those from the original dataset.
'''
```

```python
score = []
for i in range(2,11):
    cluster = KMeans(n_clusters = i)
    df["K_Means"] = cluster.fit_predict(X_new)
    score.append(silhouette_score(X_new,df["K_Means"]))
score
```

```python
#Visualization
import matplotlib.pyplot as plt
plt.scatter(range(2,11),score)
plt.plot(range(2,11),score,color="red")
plt.show()
```

```python
print("The highest silhouette score using KMeans clustering and PCA components is:",round(max(score),2))
```
