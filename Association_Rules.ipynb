```python
from apyori import apriori

import pandas as pd

# Data Preprocessing
df = pd.read_csv("C:\\Users\\sujey\\Downloads\\Assignments\\Association Rules\\Association Rules\\Online retail.csv", header = None)
df
```

```python
trans = []

for i in range(0, 7501):
  trans.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
assoc_rules = apriori(transactions = trans,min_support = 0.003,min_confidence = 0.2,min_lift = 3,min_length = 2,max_length = 2)
```

```python
final_op = list(assoc_rules)
final_op
```

```python
a=[]
b=[]
c=[]
d=[]
e=[]

for i in range(0,9):
    a.append(final_op[i][2][0][0])
    b.append(final_op[i][2][0][1])
    c.append(final_op[i][1])
    d.append(final_op[i][2][0][2])
    e.append(final_op[i][2][0][3])

a = pd.DataFrame(a)
b = pd.DataFrame(b)
c = pd.DataFrame(c)
d = pd.DataFrame(d)
e = pd.DataFrame(e)
```

```python
final_df = pd.concat([a,b,c,d,e],axis=1)
final_df.columns = ["Base","Add","Support","Confidence","Lift"]
final_df
```
